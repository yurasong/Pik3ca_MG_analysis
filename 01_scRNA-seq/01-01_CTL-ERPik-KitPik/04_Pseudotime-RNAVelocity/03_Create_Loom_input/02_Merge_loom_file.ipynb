```python
import scvelo as scv
```

```python
adata_ctl = scv.read("filtered_CTL.loom", cache=True)
adata_erpik = scv.read("filtered_ER_Pik.loom", cache=True)
adata_kitpik = scv.read("filtered_Kit_Pik.loom", cache=True)
```

```python
import anndata as ad

adata_merged = ad.concat(
    [adata_ctl, adata_erpik, adata_kitpik],
    axis=0,              # concatenate cells (rows)
    join="outer",        # keep all genes (union), fill missing with 0
    label="sample",      # add a new column in .obs to track sample origin
    keys=["CTL", "ER_Pik", "Kit_Pik"]
)
```

```python
adata_merged.obs["sample"]
```

```python
adata_merged.write("merged_Pik.h5ad")
adata_merged.write_loom("merged_Pik.loom")
```
